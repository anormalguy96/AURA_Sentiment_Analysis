```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv('../data/Reviews.csv')
print("Dataset Shape:", df.shape)
df.head()
```

```python
print("Missing values per column:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Score', data=df, palette='viridis')
plt.title('Distribution of Review Scores (1-5 Stars)')
plt.xlabel('Star Rating')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
sentiment_map = {1: 'Negative', 2: 'Negative', 3: 'Neutral', 4: 'Positive', 5: 'Positive'}
df['Sentiment'] = df['Score'].map(sentiment_map)

plt.figure(figsize=(8, 6))
sns.countplot(x='Sentiment', data=df, order=['Positive', 'Neutral', 'Negative'], palette='magma')
plt.title('Distribution of Mapped Sentiments')
plt.xlabel('Sentiment')
plt.ylabel('Number of Reviews')
plt.show()
```
